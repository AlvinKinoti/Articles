```python
#multiple line graphs in one figure (subplots)

x = np.arange(0,3,.6)
y1 = x+6
y2 = 3*x+8
y3 = 4*x-1
y4 = 5*x+2

plt.subplot(1,2,1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y4)

plt.show()
```

```python
# scatter plot

x = [2,3,4,5,6,7,8]
a = [9,8,6,4,5,2,1]
b = [6,4,7,8,2,3,5]

plt.figure(figsize= (10,10))
plt.scatter (x,a, s=300)
plt.scatter (x,b, s=600, marker = '4')
plt.xlabel('x')
plt.ylabel('a and b')
plt.title('X versus a and b')
plt.show()
```

```python
#boxplot

one = [1,2,3,4,5,6,7.8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = ([one,two,three])

plt.boxplot(data)
plt.show()
```

```python
# violin plot

one = [1,2,3,4,5,6,7.8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data = ([one,two,three])

plt.violinplot(data,  showmeans=True, showmedians = True,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y')
plt.show()
```

```python
#pie chart
 
employees =['Martin', 'Harun', 'Grace', 'Damcel', 'Greg']
performance = [230, 400, 198, 25, 214]

plt.pie(performance, labels=employees , shadow=True, explode=(0.1,0.1,0.1,0.3,0.2), autopct='%0.1f%%', 
        colors=['blue', 'red', 'yellow','grey', 'green'])
plt.show()
```

```python
#areaplot 

x = [10,20,30,40,50,60,70,80,90]
y = [13,56,32,67,87,90,31,99,35]

plt.stackplot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.show()
```

```python
#areaplot  and line graph

x = [10,20,30,40,50,60,70,80,90]
y = [13,56,32,67,87,90,31,99,35]

plt.stackplot(x,y)
plt.plot(x,y, color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.show()
```
